```python
import json
```

```python
class Aspect:
    
    def __init__(self, dataset, train):
        self.dataset = dataset
        self.datasets = {'1': {'num': 4000, 'frac_pos': 0.35, 'frac_neu': 0.35, 'frac_neg': 0.3},
                         '2': {'num': 4000, 'frac_pos': 0.6, 'frac_neu': 0.15, 'frac_neg': 0.25},
                         '3': {'num': 4000, 'frac_pos': 0.25, 'frac_neu': 0.6, 'frac_neg': 0.15},
                         '4': {'num': 4000, 'frac_pos': 0.25, 'frac_neu': 0.15, 'frac_neg': 0.6}
                        }
        self.set = 'train' if train == True else 'test'
        self.file_name = './TripAdvisor_hotel_{}_{}_{}_{}_{}.json'.\
                         format(self.datasets[dataset]['num'],
                                self.datasets[dataset]['frac_pos'],
                                self.datasets[dataset]['frac_neu'],
                                self.datasets[dataset]['frac_neg'],
                                self.set
                               )
        with open(self.file_name) as f:
            self.data = json.load(f)
        self.aspects = []
        self.aspect_rank = {}
        
    def __call__(self):
        self.count_aspect_num()
        self.count_aspect()
            
    def count_aspect_num(self):
        idx = 0
        for content in self.data:
            try:
                for aspect_term in content['opinions']['aspect_term']:
                    aspect = aspect_term['term']
                    if idx == 0:
                        self.aspects.append(aspect)
                    else:
                        if aspect not in self.aspects:
                            self.aspects.append(aspect)
                    idx += 1
            except:
                continue
        print('Number of aspects in dataset{}_{}: {}\n'.\
              format(self.dataset, self.set, len(self.aspects)))
                
    def count_aspect(self, top_k=10):
        for aspect in self.aspects:
            num = 0
            for content in self.data:
                try:
                    for aspect_term in content['opinions']['aspect_term']:
                        if aspect == aspect_term['term']:
                            num += 1
                except:
                    continue
            self.aspect_rank[aspect] = num
        # Top-K aspects
        self.aspect_rank=sorted(self.aspect_rank.items(),key=lambda x:x[1],reverse=True)[0: top_k]
        print('Top-{} aspects in dataset{}_{}:\n {}\n'.\
              format(top_k, self.dataset, self.set, self.aspect_rank))
        
    def count_avg_aspect(self):
        num_review = len(self.data)
        num_aspect = 0
        num_aspect_word = 0
        for content in self.data:
            num_aspect += len(content['opinions']['aspect_term'])
            try:
                for aspect_term in content['opinions']['aspect_term']:
                    aspect = aspect_term['term']
                    num_aspect_word += len(aspect)
            except:
                continue
        print('Average number of aspects in a sentence in dataset{0}_{1}: {2}\n' \
              'Average length of aspects in a sentence in dataset{0}_{1}: {3}\n' \
              'Number of review in dataset{0}_{1}: {4}\n' .\
              format(self.dataset, self.set, 
                     num_aspect / num_review, num_aspect_word / num_aspect,
                     num_review
                    )
             )
```

```python
datasets = ['1', '2', '3', '4']
is_trainsets = [True, False]
for dataset in datasets:
    for is_trainset in is_trainsets:
        aspect = Aspect(dataset=dataset, train=is_trainset)
        aspect()
```
